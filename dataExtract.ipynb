```python
import csv
import json
import pandas as pd

# Open the CSV file and read in the data
with open('/content/drive/MyDrive/Colab Notebooks/Country_List.csv') as data:
    reader = csv.reader(data)
    data = [row for row in reader]

# Convert the data into a dictionary
countryDict = {}
for i, row in enumerate(data):
    countryDict[row[0]] = i


# Load the CSV file into a pandas dataframe
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/gemstone15.csv')

# Group the dataframe by exporter and importer, and sum the value column
summedData = df.groupby(['Exporter', 'Importer'])['Value'].sum().reset_index()

summedData.to_csv('/content/drive/MyDrive/Colab Notebooks/summed_gem.csv', index=False)


linksdict = {}
# Open the CSV file and read in the data
with open('/content/drive/MyDrive/Colab Notebooks/summed_gem.csv') as csv1:
    reader1 = csv.DictReader(csv1)
    for row1 in reader1:
        sourceId = countryDict.get(row1['Exporter'])
        
        targetId = countryDict.get(row1['Importer'])
        if sourceId is not None and targetId is not None:
          if sourceId not in countryDict:
            linksdict[sourceId] = (targetId,float(row1['Value']))



print(linksdict)

#Convert the dictionary into a JSON object
json_data = json.dumps({"nodes": [{'id': v, 'name': k} for k, v in countryDict.items()],"links":[{'source': k, 'target': v[0],'value':v[1]} for k, v in linksdict.items()]})

# Write the JSON data to a file
with open('/content/drive/MyDrive/Colab Notebooks/countries.json', 'w') as f:
    f.write(json_data)
    print('done')


 
```
